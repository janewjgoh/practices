## Scatter Plot with Pandas, Seaborn, and Pyplot

### import libraries and sample csv 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
!wget -O FuelConsumption.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/FuelConsumptionCo2.csv
```

```python
df = pd.read_csv("FuelConsumption.csv")

# take a look at the dataset
df.head()
```

```python
df.describe(include="all")
```

### Scatter Plot

```python
# using pandas
ax = df.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS') 
# alternatively ax = df.plot(kind='scatter', x='ENGINESIZE', y='CO2EMISSIONS')

# use matplotlib artist layer to add title and labels
ax.set(title='PD Scatter Plot', xlabel='Engine Size', ylabel='CO2 Emissions')
```

```python
# using matplotlib pyplot to plot and set title, labels
plt.scatter(df.ENGINESIZE, df.CO2EMISSIONS)
plt.title('PLT Scatter Plot') 
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')
```

```python
# using seaborn sns to plot and set title
sns.scatterplot(x='ENGINESIZE', y='CO2EMISSIONS', data=df).set_title('SNS Scatter Plot')
```

```python
# put them all side by side
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(3,3,1) #(#row,#column,index)
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)

# subplot 1
sp = df.plot.scatter(x='ENGINESIZE', y='CO2EMISSIONS', ax=ax1) 
sp.set(title='PD Scatter Plot', xlabel='Engine Size', ylabel='CO2 Emissions')

# subplot 2
plt.subplot(3, 3, 2)
plt.scatter(df.ENGINESIZE, df.CO2EMISSIONS)
plt.title('PLT Scatter Plot') 
plt.xlabel('Engine Size')
plt.ylabel('CO2 Emissions')

# subplot 3
sns.scatterplot(x='ENGINESIZE', y='CO2EMISSIONS', data=df, ax=ax3).set_title('SNS Scatter Plot')
```
